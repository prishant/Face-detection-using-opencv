```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
image = cv2.imread("/content/human pic.webp")

img_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

faces = face_cascade.detectMultiScale(img_gray)
print(faces)

for (x, y, w, h) in faces :
  cv2.rectangle(image, (x,y), (x+w,y+h), (255, 255, 0), 3)
# Display the image output
cv2_imshow(image)
```

```python
eye_cascade = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
image = cv2.imread("/content/human pic.webp")
img_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
eyes = eye_cascade.detectMultiScale(img_gray)

print(eyes)

for (x,y,w,h) in eyes :
  cv2.rectangle(image, (x,y), (x+w,y+h), (255, 10, 255), 3)
# Display the image output
cv2_imshow(image)
```

```python
# function for face detection and counting
def face_detect(image_path):
  face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
  image = cv2.imread(image_path)
  img_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
  faces = face_cascade.detectMultiScale(img_gray)
  count = len(faces)
  print("The number of faces in the image : ", count)
  for (x,y,w,h) in faces :
    cv2.rectangle(image, (x,y), (x+w,y+h), (255, 255, 100), 3)
  # Display the image output
  cv2_imshow(image)
```
